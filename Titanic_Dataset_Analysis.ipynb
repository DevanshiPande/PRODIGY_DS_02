```python
#Multivariate Analysis
#Survival with Pclass
sns.countplot(x='Survived', hue='Pclass', data=df)

pd.crosstab(df['Pclass'], df['Survived']).apply(lambda r: round(r/r.sum()*100, 1), axis=1)

```

The above plot will show the count of passengers who survived and did not survive, with different colors representing different passenger classes. It helps visualize how survival rates differ across different classes

The above table depicts the code that calculates the percentage of passengers who survived and did not survive within each passenger class and rounds the percentages to one decimal place. The result is a table where each cell shows the percentage rather than the raw count, making it easier to compare survival rates across different classes.

```python
#Feature Engineering
#Creating a new column by the name of family which will be the sum of SibSp and Parch cols
df['family']=df['Parch']+df['SibSp']
```

```python
df.sample(10)
```

```python
#Create a new feature by the name of family type
def family_type(number):
  if number==0:
    return "Alone"
  elif number>0 and number<=4:
    return "Medium"
  else:
    return "Large"
```

```python
df['family_type']=df['family'].apply(family_type)
```

```python
df.sample(10)
```

```python
pd.crosstab(df['family_type'], df['Survived']).apply(lambda r: round(r/r.sum()*100, 1), axis=1)
```

Detecting Outliers based on Numerical Data and Categorical Data

```python
#Outliers detection in age(Almost Normal)
df=df[df['Age']<(df['Age'].mean()+3*df['Age'].std())]
df.shape
```

```python
#handling outliers from Fare Column
#Find quartiles
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
low_outlier = Q1 - 1.5 * (Q3 - Q1)
high_outlier = Q3 + 1.5 * (Q3 - Q1)

df = df[(df['Fare'] < low_outlier) & (df['Fare'] > high_outlier)]
```

```python
#One hot encoding
df_encoded = pd.get_dummies(data=df, columns=['Pclass', 'Sex', 'Embarked', 'family_type'], drop_first=True)
print(df_encoded.sample(4))

```

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(15, 6))
sns.heatmap(correlation_matrix, cmap='summer', annot=True)
plt.show()
```

The likelihood of female survival is higher than that of male survival.

Passengers who embarked from Cherbourg (C) had the highest survival rate, followed by those from Queenstown (Q) and Southampton (S).

Traveling in Pclass 3 was the deadliest.

People aged 20 to 40 had the lowest survival rate.

Passengers who paid lower fares had a lower chance of surviving.

People traveling with smaller families had a higher chance of surviving the accident compared to those with large families or traveling alone.
